# Extract Elements in Reverse Order:
# Write a Python function that extracts a list of elements in reverse order starting from the second-to-last element and skipping one element in between.

```python
def reverse_skip(lst: list) -> list:
  if len(lst) < 2:
      return []
  return lst[len(lst) - 2 :: -2]

# Example usage:
numbers = [1, 2, 3, 4, 5, 6]
print(f"Original list: {numbers}")
print(f"Extracted elements in reverse order: {reverse_skip(numbers)}")
```

# Write a Python function that takes a nested list and flattens it into a single list, where all the elements are in a single dimension.

```python
def flatten(lst:list) ->list:
  flat = [item for sublist in lst for item in sublist]
  return flat

input = [[1, 2], [3, 4], [5]]
print(flatten(input))
```

# Sum of All Elements in a Nested List:
# Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).

```python
def sum_nested(lst:list)->int:
  total = 0
  for element in lst:
    if isinstance(element, list):
      total += sum_nested(element)
    elif isinstance(element, (int, float)):
      total += element
  return total

input = [[1, 2], [3, [4, 5]], 6]
print(sum_nested(input))
```

# Remove Specific Element from a Nested List:
# Write a Python function that removes all occurrences of a specific element from a nested list.

```python
def remove_element(lst:list, elem:int)->list:
  result = []
  for item in lst:
    if isinstance(item, list):
      new_sublist = [x for x in item if x != elem]
      result.append(new_sublist)
    elif item != elem:
      result.append(item)
  return result

input_user_example = [[1, 2], [3, 2], [4, 5]]
print(f"Input: {input_user_example}, Element to remove: 2, Output: {remove_element(input_user_example, 2)}")
```

# 5. Find the Maximum Element in a Nested List:
# Write a Python function that finds the maximum element in a nested list (regardless of depth).

```python
def find_max(lst):
    stack = list(lst)
    max_val = float('-inf')

    while stack:
        item = stack.pop()
        if isinstance(item, list):
            stack.extend(item)
        else:
            if item > max_val:
                max_val = item
    return max_val

input_data = [[1, 2], [3, [4, 5]], 6]
print(find_max(input_data))
```

# 6. Count Occurrences of an Element in a Nested List:
# Write a Python function that counts how many times a specific element appears in a nested list.

```python
def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1  # Found a match
    return count

input_data = [[1, 2], [3, [4, 5]], 6]
print(count_occurrences(input_data, 3))
```

# 7. Flatten a List of Lists of Lists:
# Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

```python
def deep_flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result

input_data = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]
print(deep_flatten(input_data))
```

# Nested List Average:
# Write a Python function that calculates the average of all elements in a nested list.

```python
def average_nested(lst):
    total = 0
    count = 0
    for item in lst:
        if isinstance(item, list):
            sub_total, sub_count = average_nested(item)
            total += sub_total
            count += sub_count
        else:
            total += item
            count += 1
    return total, count
total, count = average_nested([[1, 2], [3, 4], [5, 6]])
avg = total / count
print(avg)
```

# 10.2 Experiment: How Fast is Numpy?
In this exercise, you will compare the performance and implementation of operations using plain Python
lists (arrays) and NumPy arrays. Follow the instructions:
1. Element-wise Addition:
• Using Python Lists, perform element-wise addition of two lists of size 1, 000, 000. Measure
and Print the time taken for this operation.
• Using Numpy Arrays, Repeat the calculation and measure and print the time taken for
this operation.


```python
import numpy as np
import time

size = 1_000_000

### 1. Using Python Lists
list_a = list(range(size))
list_b = list(range(size))

start_time = time.time()
# Element-wise addition using a list comprehension
list_result = [list_a[i] + list_b[i] for i in range(size)]
list_time = time.time() - start_time

print(f"Python List Addition Time: {list_time:.5f} seconds")


### 2. Using NumPy Arrays
array_a = np.arange(size)
array_b = np.arange(size)

start_time = time.time()
# Element-wise addition using NumPy's vectorized operation
array_result = array_a + array_b
numpy_time = time.time() - start_time

print(f"NumPy Array Addition Time:  {numpy_time:.5f} seconds")

# Calculate the speedup
print(f"\nNumPy is {list_time / numpy_time:.2f}x faster than Python lists!")
```

# Matrix Multiplication
• Using Python lists, perform matrix multiplication of two matrices of size 1000x1000. Measure and print the time taken for this operation.
• Using NumPy arrays, perform matrix multiplication of two matrices of size 1000x1000.
Measure and print the time taken for this operation.

```python
import numpy as np
import time

size = 1000

list_a = [[i for i in range(size)] for j in range(size)]
list_b = [[i for i in range(size)] for j in range(size)]
list_result = [[0 for _ in range(size)] for _ in range(size)]

start_time = time.time()
for i in range(size):
    for j in range(size):
        for k in range(size):
            list_result[i][j] += list_a[i][k] * list_b[k][j]
list_time = time.time() - start_time

print(f"Python List Matrix Multiplication Time: {list_time:.5f} seconds")

# 2. Using NumPy Arrays
array_a = np.ones((size, size))
array_b = np.ones((size, size))

start_time = time.time()
array_result = np.dot(array_a, array_b) # or array_a @ array_b
numpy_time = time.time() - start_time

print(f"NumPy Array Matrix Multiplication Time:  {numpy_time:.5f} seconds")

speedup = list_time / numpy_time
print(f"\nNumPy is {speedup:.2f}x faster at matrix multiplication!")
```
